```python
import pandas as pd
df = pd.read_csv('../notebooks/Myocardial infarction complications Database (1).csv')
df.head()
```

```python
lista_de_delete = ["ID", "IBS_NASL", "KFK_BLOOD", "JELUD_TAH", "FIBR_JELUD", "A_V_BLOK", "OTEK_LANC", "RAZRIV", "DRESSLER", "REC_IM", "P_IM_STEN", "LET_IS", "PREDS_TAH", "FIBR_PREDS"]
df = df.drop(lista_de_delete, axis=1)
df.head(1700)
```

Criação de um dicionário apenas para saber qual o index de cada coluna.

```python
columns_list = df.columns.tolist()

dicionario = {}
for coluna in columns_list:
    index = columns_list.index(coluna)
    dicionario[coluna] = index
print(dicionario)
```

```python

tirar_a_moda = df.columns

for coluna in tirar_a_moda:
    if coluna == "AGE":
        mediana = df[coluna].median()
        df[coluna].fillna(mediana, inplace=True)
    elif coluna == "SEX" or coluna == "S_AD_KBRIG" or coluna == "D_AD_KBRIG" or coluna == "GIPO_K" or coluna == "K_BLOOD" or coluna == "GIPER_Na" or coluna == "Na_BLOOD" or coluna == "ALT_BLOOD" or coluna == "AST_BLOOD" or coluna == "ROE" or coluna == "TIME_B_S" or coluna == "NA_KB" or coluna == "NOT_NA_KB" or coluna == "LID_KB" or coluna == "R_AB_3_n" or coluna == "NITR_S":
        continue
    elif coluna == "S_AD_ORIT" or coluna == "D_AD_ORIT":
        media = df[coluna].median()
        df[coluna].fillna(mediana, inplace=True)
    else:
        moda = df[coluna].mode().values[0]
        df[coluna].fillna(moda, inplace=True)
print("FIM DO LOOP\n")
```
