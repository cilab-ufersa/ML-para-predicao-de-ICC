Importando o Pandas e lendo o arquivo.csv.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Myocardial infarction complications Database.csv')
df.head(1700)
```

Deletando colunas com cerca de 30% atÃ© 90% de valores vazios e colunas referentes a outros tipos de complicaÃ§Ãµes que nÃ£o sÃ£o o nosso foco no momento.

```python
delete_list = ["ID", "IBS_NASL", "KFK_BLOOD", "JELUD_TAH", "FIBR_JELUD", "A_V_BLOK", "OTEK_LANC", "RAZRIV", "DRESSLER", "REC_IM", "P_IM_STEN", "LET_IS", "PREDS_TAH", "FIBR_PREDS", "D_AD_KBRIG", "S_AD_KBRIG", "GIPO_K", "NA_KB", "LID_KB"]
df = df.drop(delete_list, axis=1)
```

CriaÃ§Ã£o de um dicionÃ¡rio apenas para saber qual o index de cada coluna, caso seja necessÃ¡rio.

```python
columns_list = df.columns.tolist()

index_dictionary = {}
for column in columns_list:
    index = columns_list.index(column)
    index_dictionary[column] = index
print(index_dictionary)
```

Criando uma funÃ§Ã£o para saber se as colunas estÃ£o com valores vazios ou nÃ£o.

```python
def verify_is_nan(df, column):
    is_nan = df[column].isna().any().any()
    if is_nan:
        print(f"Existem valores nulos na coluna {column}.\n")
    else:
        print(f"NÃ£o existem valores nulos na coluna {column}.\n")
```

Nesse trecho de cÃ³digo abaixo, estÃ¡ sendo aplicado uma tratativa de dados, preenchendo valores vazios por valores modais, medianos ou medios.

```python

columns_list = df.columns

for column in columns_list:
    if column == "AGE" or column == "S_AD_ORIT" or column == "D_AD_ORIT" or column == "L_BLOOD" or column == "ROE" or column == "NA_BLOOD":
        median_columns = df[column].median()
        df[column].fillna(median_columns, inplace=True)
        df[column] = df[column].astype('int32')
        verify_is_nan(df, column)
    elif column == "AST_BLOOD" or column == "ALT_BLOOD" or column == "K_BLOOD" :
        mean_columns = df[column].mean()
        df[column].fillna(mean_columns, inplace=True)
        df[column] = df[column].astype('float64')
        verify_is_nan(df, column)
    elif column == "SEX":
        continue
    else:
        mode_columns = df[column].mode().values[0]
        df[column].fillna(mode_columns, inplace=True)
        df[column] = df[column].astype('int32')
        verify_is_nan(df, column)
        
```
