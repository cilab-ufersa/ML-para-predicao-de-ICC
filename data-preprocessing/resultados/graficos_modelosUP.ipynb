```python
import pandas as pd
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
y_real = pd.read_csv('dataset_tratado.csv')
y_pred_RFUP = pd.read_csv('C:\\Users\\angel\\Cilab\\projeto\\tratativa_de_dados_IM\\data-preprocessing\\models\\random_forest_models\\predicoes_RFUP.csv')
y_pred_DTUP = pd.read_csv('C:\\Users\\angel\\Cilab\\projeto\\tratativa_de_dados_IM\\data-preprocessing\\models\\decision_tree_models\\predicoes_DTUP.csv')
y_pred_LGBMUP = pd.read_csv('C:\\Users\\angel\\Cilab\\projeto\\tratativa_de_dados_IM\\data-preprocessing\\models\\lightGBM_models\\predicoes_LGBMUP.csv')
```

```python
fpr, tpr, _ = metrics. roc_curve(y_pred_RFUP['y_real'].values, y_pred_RFUP['y_pred'].values)
auc = metrics.roc_auc_score(y_pred_RFUP['y_real'].values, y_pred_RFUP['y_pred'].values)
plt.plot(fpr,tpr,label="RandomForest, auc="+str(auc))
plt.legend(loc=4)

fpr, tpr, _ = metrics. roc_curve(y_pred_DTUP['y_real'].values, y_pred_DTUP['y_pred'].values)
auc = metrics.roc_auc_score(y_pred_DTUP['y_real'].values, y_pred_DTUP['y_pred'].values)
plt.plot(fpr,tpr,label="DecisionTree, auc="+str(auc))
plt.legend(loc=4)

fpr, tpr, _ = metrics. roc_curve(y_pred_LGBMUP['y_real'].values, y_pred_LGBMUP['y_pred'].values)
auc = metrics.roc_auc_score(y_pred_LGBMUP['y_real'].values, y_pred_LGBMUP['y_pred'].values)
plt.plot(fpr,tpr,label="LightGBM, auc="+str(auc))
plt.legend(loc=4)

plt.show()
```
